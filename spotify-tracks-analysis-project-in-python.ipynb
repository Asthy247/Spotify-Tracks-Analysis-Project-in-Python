Each row represents a song with the following information:

**id**: A unique identifier for the song.

**name:** The title of the song.

**popularity**: A measure of the song's popularity on Spotify.

**duration_ms**: The duration of the song in milliseconds.

**explicit**: Whether the song contains explicit lyrics (1) or not (0).

**artists**: A list of artists associated with the song.

**id_artists**: A list of artist IDs.

**release_date**: The release date of the song.

**danceability:** Danceability describes how suitable a track is for dancing.

**energy**: Energy is a measure of a track's intensity and activity.

**key**: The key of the track.

****loudness:** **The overall loudness of a track in decibels (dB).

**speechiness:** Speechiness detects the presence of spoken words in a track.

**acousticness**: A confidence measure of whether the track is acoustic.

**instrumentalness**: The presence of vocals in a track.  

**liveness:** Detects the presence of a live audience.

**valence**: A measure of musical positiveness.

**tempo:** The overall tempo of a track in beats per minute (BPM).

**time_signature:** The time signature of a track.

# Data Visualization

# Histogram of Song Durations

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Song Durations
plt.figure(figsize=(10, 6))
sns.histplot(tracks['duration'], bins=30, kde=True)
plt.title('Distribution of Song Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Count')
plt.show()
```

The peak around 200 seconds suggests that users might prefer songs of this duration. 


This could be influenced by factors like attention span, listening habits, and the structure of popular music genres.

# Correlation Map

```python
# unwanted columns are dropped
# "pearson" correlation method is applied
# set figure size to 14,6
# use seaborn heatmap function to create correlation map
# setting annotation = True, this will write the data value in each cell.
# set fmt = .1g (this is a string formating tool to use when adding annotation)
# vmin and vmax are the values to add for color map
# cmap - colormap
```

```python
# Drop the specified columns
tracks_subset = tracks.drop(["key", "mode", "explicit"], axis=1)

# Convert the entire DataFrame to numeric
tracks_numeric = tracks_subset.apply(pd.to_numeric, errors='coerce')

# Calculate correlation
correl = tracks_numeric.corr(method='pearson')

# Now, proceed with your visualization
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(correl, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor="Black")
heatmap = plt.title("Correlation HeatMap Between Variable")
```

**Positive Correlations:**

o**Popularity and Duration**n: There's a weak positive correlation, suggesting that longer songs tend to be more popular


**Negative Correlations:**

**Acousticness and Energy:** There's a strong negative correlation, indicating that acoustic songs tend to have lower energy levels..


# Data Sampling

```python
# we are going to sample our data in just 0.4% of total data.
# will create 2 regression plot using this data
```

```python
sample_df = tracks.sample(int(0.004*len(tracks)))

# use int function. 
# 0.004 multiplied by len of original data frame
```

```python
print(len(sample_df))

# will print the length of my sample data frame
# if i run this, o.4% of my total dataset is 2346
```

# Regression Plot between Loudness and Energy

```python
# now we'll create a regression plot between loudness and energy

plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")

# I'll give color to data variable. Let's say color is C
```

# Regression Plot for Popularity and Accousticness

```python
# Now we'll make another regression plot for popularity and accousticness.

plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "b").set(title = "Popularity vs Acousticness Correlation")


# when i run this,I have different points for songs
# the regression line is downwards, which means acousticness of song increases when popularity decreases and vice versa.
```
